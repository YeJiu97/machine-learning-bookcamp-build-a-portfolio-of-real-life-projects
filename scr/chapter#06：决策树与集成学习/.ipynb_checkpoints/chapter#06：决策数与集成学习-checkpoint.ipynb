````python
# 这里开始进行转换
status_values = {
    1: 'ok',  # 1表示没有问题
    2: 'default',  # 2表示违约
    0: 'unk'  # 0表示unknown，也就是值缺失
}
````

# 决策树

决策树是一种基于树结构的机器学习模型，用于解决分类和回归问题。它通过一系列的决策节点和叶子节点来表示数据的分类规则或者数值预测规则。

在决策树中，每个内部节点表示一个特征/属性的判断条件，每个分支代表了该属性的一个取值，而每个叶子节点代表了一个类别标签（用于分类问题）或者一个数值（用于回归问题）。

决策树的构建过程通常包括以下步骤：

1. **选择划分属性：** 从所有属性中选择一个最优的划分属性，以将数据集划分为不同的子集。通常使用一些指标（比如信息增益、基尼指数等）来评估每个属性的划分能力。
   
2. **划分数据集：** 根据选择的划分属性将数据集划分为不同的子集。每个子集对应于划分属性的不同取值。
   
3. **递归构建子树：** 对每个子集递归地重复上述过程，直到满足某种停止条件（比如达到最大深度、节点样本数小于某个阈值等）。
   
4. **生成决策树：** 重复以上步骤直到构建完整的决策树。

决策树模型具有直观、易解释的特点，能够处理混合型数据、不需要对数据进行过多的预处理，而且对异常值不敏感。它们通常用于解决分类和回归问题，并且可以用于特征选择、数据探索和决策推理等任务。

这是一个示例：



````python
# 这是上面的例子的代码
def assess_risk(client):
    if client['records'] == 'yes':
        if client['job'] == 'parttime':
            return 'default'
        else:
            return 'ok'
    else:
        if client['assets'] > 6000:
            return 'ok'
        else:
            return 'default'
````

因为需要使用决策树，所以需要从tree包中导入：

````python
from sklearn.tree import DecisionTreeClassifier  # 导入对饮的包

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
````

接着对结果进行评估，使用AUC作为指标：

````python
from sklearn.metrics import roc_auc_score

y_pred = dt.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, y_pred)
````

结果显示的是，模型在训练集上实现了100%完美的预测。

但是我们需要注意的是：在训练集上的100%完美预测既有可能是因为模型的性能特别的好，也有可能是因为模型在训练集上出现了过拟合的情况。

````python
# 使用验证集来进行测试
y_pred = dt.predict_proba(X_val)[:, 1]
roc_auc_score(y_val, y_pred)
````

可以发现，在验证集上，只有65的准确率。

可以发现，原先的模型是一个典型的过拟合的情况，这是决策树本身的原理决定的：



决策树可以不断地增加节点和深度，从而为每一个样本都给出一个最终节点，从而完美完美的分类。

为了防止过拟合，一个措施是现实树的深度。

````python
# 默认情况之下，树的深度参数为None
dt = DecisionTreeClassifier(max_depth = 2)
dt.fit(X_train, y_train)
````

可以使用tree包中的export_text函数来进行可视化：

````python
from sklearn.tree import export_text

tree_text = export_text(dt, feature_names=dv.feature_names_)
print(tree_text)
````

这个结果是一个简单的决策树模型的输出，描述了一棵决策树的结构和决策规则。让我来解释一下：

- 树的每一层都包含一个条件判断，称为节点。
- 从根节点开始，根据条件判断的结果，选择不同的分支，直到到达叶子节点，叶子节点给出最终的分类结果。

这个决策树包含了两个特征（`records` 和 `seniority`）和一个目标变量（`class`）。

- 如果 `records` 的取值小于等于 0.50，则进入左子树，否则进入右子树。
  - 如果 `records` 的取值小于等于 0.50，并且 `job` 的取值小于等于 0.50，则预测结果为 `False`。
  - 如果 `records` 的取值小于等于 0.50，并且 `job` 的取值大于 0.50，则预测结果为 `True`。
- 如果 `records` 的取值大于 0.50，则进入右子树。
  - 如果 `seniority` 的取值小于等于 6.50，则预测结果为 `True`。
  - 如果 `seniority` 的取值大于 6.50，则预测结果为 `False`。

这个决策树描述了如何根据输入特征 `records` 和 `seniority` 来预测目标变量 `class`。通过观察不同特征的取值，树可以按照给定的规则对数据进行分类。

或者可以使用这张图来进行观察：



可也可以使用其他的可视化方法：

````python
from sklearn import tree

plt.figure(figsize=(10, 6))
tree.plot_tree(dt, feature_names=dv.feature_names_, filled=True, rounded=True, fontsize=10)
plt.show()
````

接着看一下这个决策树的性能如何：

````python
y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred)
print('train auc', auc)

y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print("validation auc", auc)
````

从结果来看，可以发现决策树在训练集上的准确度下降了，但是在验证集上的准确性上升了。

但是这个树太简单了，所以需要尝试不同的参数并且查看哪些参数的效果最佳，能够带来最好的AUC。

````python
# 尝试不同的参数，并且计算对应的auc分数
for depth in [1, 2, 3, 4, 5, 6, 10, 15, 20, None]:
    dt = DecisionTreeClassifier(max_depth = depth)
    dt.fit(X_train, y_train)
    y_pred = dt.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    print('%4s -> %.3f' % (depth, auc))
````

可以发现当max_depth的值为5的情况之下是最好的。

但是我们需要采用4，5，6三个max_depth，然后调试一下min_samples_leaf的参数：

````python
for m in [4, 5, 6]:
    
    print('depth: %s' % m)

    for s in [1, 5, 10, 15, 20, 50, 100, 200]:
        dt = DecisionTreeClassifier(max_depth = m, min_samples_leaf = s)
        dt.fit(X_train, y_train)
        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        print('%4s -> %.3f' % (s, auc))
````

结果最好的是depth=6和min_samples_leaf=15。

````python
# 使用这个最佳的参数来训练最后的树
dt = DecisionTreeClassifier(max_depth = 6, min_samples_leaf=15)
dt.fit(X_train, y_train)
````

````python
plt.figure(figsize=(20, 12))
tree.plot_tree(dt, feature_names=dv.feature_names_, filled=True, rounded=True, fontsize=10)
plt.show()
````

可视化不是非常的美观，因为存在着一些重叠。

````python
plt.figure(figsize=(40, 24))
tree.plot_tree(dt, feature_names=dv.feature_names_, filled=True, rounded=True, fontsize=10)
plt.show()
````

这样区分度就大了很多，但是文字太小了。

# 随机森林

随机森林是一种流行的机器学习算法，它属于集成学习方法。随机森林通过构建多个决策树来做出预测，其核心思想是“集思广益”，通过合并多个模型的预测结果来提高整体的预测准确性和稳定性。这里是随机森林的几个关键特点：

1. **构建多个决策树**：随机森林通过在训练数据的不同子集上训练多个决策树来工作。每棵树的训练数据都是通过从原始数据集中进行有放回抽样得到的，这种方法称为自助聚合（bootstrap aggregating 或 bagging）。

2. **特征选择的随机性**：在构建决策树的每个分裂点时，算法会从所有可用的特征中随机选择一个特征子集，并从中找出最佳分裂特征。这一步增加了模型的多样性，有助于提高模型的泛化能力。

3. **预测决策**：对于分类问题，随机森林的预测结果是基于所有决策树预测结果的多数投票。对于回归问题，预测结果通常是所有决策树预测结果的平均值。

随机森林的优点包括：
- **鲁棒性**：由于训练多个树并结合它们的预测，随机森林通常对噪声和异常值不太敏感。
- **减少过拟合**：随机选择特征和训练样本的方法可以有效减少模型过拟合的风险。
- **易于使用**：随机森林算法对数据预处理的需求较低，且参数调整相对简单。

不过，随机森林也有缺点，比如模型可能比单个决策树复杂，需要更多的计算资源，且模型的可解释性不如单棵决策树。

````python
# 导入随机森林模型
from sklearn.ensemble import RandomForestClassifier
````

第一件事情是指定集成中树的数量，可以使用n_estimators参数：

````python
rf = RandomForestClassifier(n_estimators=10)  # n_estimators = 10，集成树中有10颗树
rf.fit(X_train, y_train)  # 开始进行训练
````

开始对结果进行AUC评分：

````python
y_pred = rf.predict_proba(X_val)[:, 1]
roc_auc_score(y_val, y_pred)
````

这里的输出范围结果为：78%。但是，这个结果可能会发生改变，每次重新训练模型，这个结果的数值都可能会产生一些变化。

````python
# 限制随机化，从而确保每次运行得到的结果是一致的
rf = RandomForestClassifier(n_estimators = 10, random_state = 3)  # random_state = 3，设置随机数种子
rf.fit(X_train, y_train)  # 开始进行训练
````

````python
# 开始进行评估
y_pred = rf.predict_proba(X_val)[:, 1]
roc_auc_score(y_val, y_pred)
````

这次得到了78%的值，并且无论再训练多少次，这个值也不会发生改变了。

````python
# 修改集成树中的树的数量
aucs = []  # 创建一个列表来存放auc的计算结果

for i in range(10, 201, 10):
    # 指定模型和进行训练
    rf = RandomForestClassifier(n_estimators=i, random_state=3)
    rf.fit(X_train, y_train)
    # 进行预测和计算auc的分数
    y_pred = rf.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    # 打印计算的结果
    print('%s -> %.3f' % (i, auc))
    # 存储auc的值
    aucs.append(auc)
````

观察数值的结果，可以认为最高的auc的值为0.820，基于奥卡姆的剃刀原则，选择n_estimators=90是一个最佳的选择。

````python
# 进行绘制
plt.plot(range(10, 201, 10), aucs)
````

绘图则指出，在树的数量达到了25~30左右的时候，模型的性能快速的增长然后开始放缓。在数量为130之后，添加更多的树将不再有帮助：模型性能大概维持在82%左右的水平。

但是数的数量不是唯一可以调整的参数，还可以调整max_depth和min_samples_leaf的值。

````python
all_aucs = {}  # 这里创建一个字典

for depth in [5, 10, 20]:  # 遍历不同的深度
    print('depth: %s' % depth)
    # 对当前的深度进行探索
    aucs = []
    for i in range(10, 201, 10):  # 尝试不同的树的数量的集成树
        rf = RandomForestClassifier(n_estimators = i, max_depth = depth, random_state = 1)  # 对应的参数
        rf.fit(X_train, y_train)  # 开始进行训练
        y_pred = rf.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        print('%s -> %.3f' % (i, auc))
        aucs.append(auc)
    all_aucs[depth] = aucs
    print()
````

可以绘制一下折线图：

````python
num_trees = list(range(10, 201, 10))
plt.plot(num_trees, all_aucs[5], label = 'depth = 5')
plt.plot(num_trees, all_aucs[10], label = 'depth = 10')
plt.plot(num_trees, all_aucs[20], label = 'depth = 20')
plt.legend()
````

这里还可以加上min_samples_leaf，只需要添加一个循环：

````python
all_aucs = {}

for depth in [5, 10, 20]:  # 遍历不同的深度
    print('Depth: %s' % depth)
    depth_dict = {}  # 为每个深度创建一个字典来存储min_samples_leaf的结果
    for min_samples_leaf in [1, 5, 10]:  # 遍历不同的min_samples_leaf值
        print('  Min samples leaf: %s' % min_samples_leaf)
        aucs = []
        for n_estimators in range(10, 201, 10):  # 遍历不同的n_estimators值
            rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=depth, min_samples_leaf=min_samples_leaf, random_state=1)
            rf.fit(X_train, y_train)
            y_pred = rf.predict_proba(X_val)[:, 1]
            auc = roc_auc_score(y_val, y_pred)
            print('    %s trees -> AUC: %.3f' % (n_estimators, auc))
            aucs.append(auc)
        depth_dict[min_samples_leaf] = aucs  # 将当前min_samples_leaf的结果列表存储在深度字典中
    all_aucs[depth] = depth_dict  # 将深度字典存储在主字典中
    print()
````

也可以这么写：

````python
all_aucs = {}
 
for m in [3, 5, 10]:
    print('min_samples_leaf: %s' % m)
    aucs = []
 
    for i in range(10, 201, 20):
        rf = RandomForestClassifier(n_estimators=i, max_depth=10, min_samples_leaf=m, random_state=1)
        rf.fit(X_train, y_train)
        y_pred = rf.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        print('%s -> %.3f' % (i, auc))
        aucs.append(auc)
   
    all_aucs[m] = aucs
    print()
````

````python
# 可视化
num_trees = list(range(10, 201, 20))
plt.plot(num_trees, all_aucs[3], label='min_samples_leaf=3')
plt.plot(num_trees, all_aucs[5], label='min_samples_leaf=5')
plt.plot(num_trees, all_aucs[10], label='min_samples_leaf=10')
plt.legend()
````

现在得到的结果为：
- 树的数量为200，n_estimators = 200
- max_depth = 10
- min_samples_leaf = 5

````python
rf = RandomForestClassifier(n_estimators = 200, max_depth = 10, min_samples_leaf= 5, random_state =1)
rf.fit(X_train, y_train)
````

# 梯度提升

梯度提升（Gradient Boosting）是一种强大的机器学习技术，用于回归和分类问题，属于集成学习方法中的提升（Boosting）类算法。其核心思想是逐步构建模型，通常是决策树，每一步都在尝试减少前一步的残差，通过这种方式，模型逐渐变得更加强大和精确。这里是梯度提升的几个关键特点和工作机制：

**关键特点**

1. **迭代学习**：梯度提升通过逐步添加新的模型，尝试纠正前一模型的错误。每一步添加的新模型是基于梯度下降算法来最小化损失函数。

2. **损失函数优化**：在每一步，算法都计算损失函数的梯度（即误差的方向和大小），新模型的任务是预测这个梯度，从而改进整体模型的表现。

3. **弱学习器**：梯度提升通常使用简单的模型作为基学习器，如深度较浅的决策树。这些基学习器通常表现不会非常好，但组合起来可以形成一个非常强大的模型。

4. **正则化技术**：梯度提升包括几种正则化手段来避免过拟合，如限制树的数量、树的深度、学习速率和子采样等。

**工作机制**

1. **初始化**：梯度提升首先使用一个基准预测器开始，通常是数据的平均值或使用一个简单的模型。

2. **循环构建**：然后，根据当前模型的残差计算损失函数的梯度。构建一个新的模型来预测这些梯度，然后将其加入到现有的模型集合中。

3. **更新预测**：通过结合现有模型和新模型的预测来更新整体预测。这个过程循环进行，直到达到指定的迭代次数或直到进一步改进变得微不足道。

4. **输出最终模型**：经过多轮迭代后，组合所有的弱学习器形成最终模型。

梯度提升在多个领域都有广泛应用，因为它通常能提供高精度的解决方案，特别是在结构化数据的预测问题中，如竞赛平台Kaggle上的许多竞赛。常见的梯度提升实现包括 XGBoost、LightGBM 和 CatBoost，这些都是优化了的版本，提供了更快的训练速度和更好的可扩展性。

原始参数解释和控制过程：

- eta（学习率）：控制每一步的学习速度。如果设置得太高，模型可能会过快收敛到次优解。
- max_depth（最大深度）：决定树可以生长的最大深度，类似于决策树分类器中的 max_depth 参数。
- min_child_weight（最小子节点权重）：决定最小的叶子节点样本权重之和，相当于决策树分类器中的 min_leaf_size。
- objective（目标函数）：定义模型要解决的问题类型，对于分类问题，应设置为 binary:logistic。
- eval_metric（评估指标）：用于评估模型性能的指标，例如可以是 "AUC"。
- nthread（线程数）：定义用于训练模型的线程数。XGBoost 可以利用多线程来加速训练。
- seed（种子）：用于生成随机数，确保结果的可重复性。
- silent（静默模式）：控制输出的详细程度。如果设置为“1”，则不输出任何警告或信息。

**找到了最佳模型参数只要还要用训练数据集和验证数据集合并之后的数据进行重新训练？**

这是一个很好的问题，涉及到机器学习中的一个关键概念：**数据利用的最大化**和**模型泛化能力的提升**。虽然通过训练数据集训练和通过验证数据集找到了最佳参数，您可能已经开发出了一个性能不错的模型，但是还有几个原因可能会促使您用合并后的数据再次进行训练：

1. **更多的训练数据**：使用更多的数据来训练模型通常会提高模型的泛化能力。当您将训练数据集和验证数据集合并后，模型可以从更多样本中学习，这可能帮助模型更好地捕捉数据的底层模式，从而在未见过的测试数据上表现更好。

2. **最优参数的再验证**：虽然您在原始的训练集和验证集上找到了最佳参数，但这些参数是在一定数据规模下得出的。当数据规模增加，相同的参数可能不再是最佳选择。重新训练可以帮助确认这些参数是否仍然适用于更大规模的数据。

3. **防止过拟合验证集**：在机器学习模型开发过程中，有时可能无意中对验证集进行了“过拟合”。这是因为模型参数的选择是为了在验证集上达到最佳性能。通过合并数据集并重新训练，您可以减少对原始验证集过度优化的风险。

4. **充分利用有限数据**：在数据资源有限的情况下，为了构建最强的可能模型，您需要利用可用的每一点数据。合并训练集和验证集重新训练可以在不泄露测试集信息的前提下，最大化训练数据的使用。

综上所述，虽然找到了最佳参数，但重新使用所有可用的非测试数据进行训练，可以在不牺牲测试的独立性和公正性的情况下，尝试进一步提升模型的性能。这种做法尤其在数据较少或者模型复杂度高的情况下显得更加重要。
